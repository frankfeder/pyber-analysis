```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import statistics
```

```python
pyber_ride_df = pd.read_csv("./Resources/PyBer_ride_data.csv")
pyber_ride_df.head()
```

```python
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind="bar")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])
x_axis = np.arange(len(pyber_ride_df))
xtick_locations = [value for value in x_axis]
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind="bar", yerr=stdev, capsize=3)
plt.xticks(xtick_locations, pyber_ride_df["Month"], rotation = 0)
plt.yticks((np.arange(0, 51, step=5.0)))
plt.show()
```
